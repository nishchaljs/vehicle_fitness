```python
resul = ['petrol','diesel']
import pandas as pd
import Levenshtein

def closest_result(word):
    df = pd.DataFrame({'result': ['petrol','diesel']})
    df['dist'] = df['result'].apply(lambda month: Levenshtein.distance(word, month))
    idxmin = df['dist'].idxmin()
    return [idxmin+1, df.loc[idxmin, 'dist']]
```
